# **First two layers of patches generated from Soneira peebles model**

##### **Packages**

```python
import matplotlib.pyplot as plt   ### data visualization package
import numpy as np
```

##### **Parameters**

```python
R = 1     ### radius of first layer
eta = 3   ### number of patches in a circle
df = 1.2  ### fractal dimension
```

##### **Algorithm**

```python
ax = plt.gca()

### first layer
main_circle = plt.Circle((0,0),radius=R,color='#7AC5CD',alpha=0.5)
ax.add_patch(main_circle)


for i in range(0,eta):
  ### patches on first layer
  angle = np.random.uniform(0, 2 * np.pi)
  radius = np.random.uniform(0, R)
  x = radius * np.cos(angle)
  y = radius * np.sin(angle)

  ax.scatter(x,y,color='blue')
  ax.text(x,y,'1')

  ### second layer
  sub_circle = plt.Circle((x,y),radius=R/(eta**(1/df)),color='#FF7256',alpha=0.5)
  ax.add_patch(sub_circle)

  for j in range(0,eta):
    ### patches on second layer
    angle = np.random.uniform(0, 2 * np.pi)
    radius = np.random.uniform(0, R / (eta**(1/df)))
    new_x = x + radius * np.cos(angle)
    new_y = y + radius * np.sin(angle)

    ax.scatter(new_x,new_y,color='red')
    ax.text(new_x,new_y,'2')

plt.axis('scaled')
plt.title("Soneira-Peebless model")
plt.xlabel('x coordinate')
plt.ylabel('y coordinate')
plt.show() 
```
